```python
import automech as am

name_dct = {
    "CCCO[O]": "RO2",
    "C[CH]COO": "QOOH",
    # bimolecular 1
    "CC[CH2]": "R",
    "[O][O]": "O2",
    # bimolecular 2
    "CC1CO1": "cyC3e12m1",
    "[OH]": "OH",
}

mech = am.from_smiles(
    rxn_smis=[
        "[O][O].CC[CH2]>>CCCO[O]",
        "CCCO[O]>>C[CH]COO",
        "C[CH]COO>>CC1CO1.[OH]",
    ],
    name_dct=name_dct,
)
am.display(mech)
```

```python
am.io.write(mech, "mech.json")
```

```python
import automol

for smi, name in name_dct.items():
    geo = automol.smiles.geometry(smi)
    print(f"{name = }")
    print(automol.geom.xyz_string(geo))
    print()
```
