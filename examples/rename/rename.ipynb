```python
from pathlib import Path

import automech
```

```python
spc_df = automech.io.rmg.read.species(inp=Path("orig_species.txt"))
```

```python
spc_df
```

```python
spc_df0 = spc_df.unique(subset=["amchi", "spin", "charge"])
```

```python
source_mech = automech.io.read(Path("full_parent.json"))
spc_df_ = source_mech.species
```

```python
import polars as pl

species_data0 = pl.read_csv("species.csv")
species_data = species_data0.with_columns(
    pl.col("name")
    .replace(spc_df_["name"], spc_df_["amchi"])
    .replace(spc_df0["amchi"], spc_df0["name"])
)
species_data.write_csv("updated_species.csv")
```
