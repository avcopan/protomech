```python
import automech
```

```python
name_dct = {
    "O": "H2O",
    "[OH]": "OH",
    "CC(O)C": "C3o2",
    "CC(O)[CH2]": "C3o2y1",
}
mech = automech.from_smiles(
    spc_smis=list(name_dct.keys()),
    rxn_smis=["CC(O)C.[OH]>>CC(O)[CH2].O"],
    name_dct=name_dct,
)
mech, *_ = automech.expand_stereo(mech, enant=True)
```

```python
mech
```

```python
automech.io.mechanalyzer.write.mechanism(
    mech, rxn_out="mechanism.dat", spc_out="species.csv"
)
```
